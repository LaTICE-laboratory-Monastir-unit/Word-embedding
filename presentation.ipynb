# Word2Vec
* Transforme les mots vers des représentations vectorielles continues (c'est-à-dire des points dans un espace de dimension N)
* Apprend les vecteurs à partir des données text (apprentissage non supervisé)

## Comment il marche?

```python
Image(filename='./images/img2.png')
```

* Deux modèles distincts
    * Skip-gram.
    * CBOW :continuous bag of words .

Étant donné une taille de fenêtre de n mots autour d'un mot w :
- Le modèle **skip-gram** prédit les mots voisins Étant donné le mot actuel. 

En revanche, 

- Le modèle **CBOW** prédit le mot actuel w, Étant donné des mots voisins dans la fenêtre.

# Skip-gram
Skip-gram typically outperforms CBOW on semantic and syntactic accuracy (Mikolov et al.)

```python
class W2vMeanVectorizer(object):
    def __init__(self, word2vec):
        self.word2vec = word2vec
        # if a text is empty we should return a vector of zeros
        # with the same dimensionality as all the other vectors
        self.dim = 100

    def fit(self, X, y):
        return self
    def fit_transform(self, X):
        return np.array([
            np.mean([self.word2vec[w] for w in words if w in self.word2vec]
                    or [np.zeros(self.dim)], axis=0)
            for words in X
        ])
    
    def transform(self, X):
        return np.array([
            np.mean([self.word2vec[w] for w in words if w in self.word2vec]
                    or [np.zeros(self.dim)], axis=0)
            for words in X
        ])
```

```python
import os
import re
from nltk.tokenize import word_tokenize
from gensim.models.doc2vec import TaggedDocument
from gensim.models import Doc2Vec
```

### Preprocessing methodes

```python
def preprocessing(string):
    string =re.sub(r'[^\u0600-\u06FF]', ' ', string)  
    return re.sub(r"\s{2,}", " ", string).strip()
```

### Helper class for momery

```python
class LabeledLineSentence(object):
    def __init__(self, filename):
        self.filename = filename
    
    def __iter__(self):
        for uid, line in enumerate(open(self.filename,'r',encoding='utf8')):
            yield TaggedDocument(word_tokenize(line), ['SENTENCE_%s' % uid])  
```

```python
### Create and train  doc2vec model 
```

```python
sentences = LabeledLineSentence('test1.txt')
model = Doc2Vec(alpha=0.025, min_alpha=0.025,size=100,min_count=3,workers=4)  # use fixed learning rate
model.build_vocab(sentences)
for epoch in range(1):
    model.train(sentences,total_examples=model.corpus_count,epochs=epoch)
    model.alpha -= 0.002  # decrease the learning rate
    model.min_alpha = model.alpha  # fix the learning rate, no decay
```

```python
### Save and load the model
```

```python
model_name='./saved_models/model'
model.save(model_name)
doc2vec = Doc2Vec.load(model_name)  
```

```python
class D2vVectorizer(object):
    def __init__(self, doc2vec):
        self.doc2vec = doc2vec
        self.dim = 100

    def fit(self, X, y):
        return self
    def fit_transform(self, X):
        return np.array([doc2vec.infer_vector(sentence) for sentence in X  ])
    
    def transform(self, X):
        return np.array([doc2vec.infer_vector(sentence) for sentence in X  ])
```
